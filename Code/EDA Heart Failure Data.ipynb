```r
install.packages('KMsurv')
install.packages('survminer')
library(KMsurv)
library(survival)
library(ggplot2)
library(survminer)
```

# Import Dataset

```r
heart1 <- read.csv('processed_heart.csv')

heart1$anaemia <- as.factor(heart1$anaemia)
heart1$creatinine_phosphokinase <- as.factor(heart1$creatinine_phosphokinase)
heart1$diabetes <- as.factor(heart1$diabetes)
heart1$ejection_fraction <- as.factor(heart1$ejection_fraction)
heart1$high_blood_pressure <- as.factor(heart1$high_blood_pressure)
heart1$platelets <- as.factor(heart1$platelets)
heart1$serum_creatinine <- as.factor(heart1$serum_creatinine)
heart1$serum_sodium <- as.factor(heart1$serum_sodium)
heart1$sex <- as.factor(heart1$sex)
heart1$smoking <- as.factor(heart1$smoking)

head(heart1)
```

## KM-Curve

```r
splots <- list()

variables <- setdiff(colnames(heart1), c("time", "DEATH_EVENT"))

for (var_name in variables[2:11]) {
  var_name <- as.character(var_name)
  fit <- survfit(as.formula(paste("Surv(time, DEATH_EVENT) ~", var_name)), data=heart1)
  plot <- ggsurvplot(pval = TRUE, fit, data=heart1, xlab="Days", ggtheme=theme_minimal(), title=paste('KM-Curve for', var_name))
  splots[[var_name]] <- plot
}

options(repr.plot.width = 25, repr.plot.height = 15, repr.plot.res = 100)

arrange_ggsurvplots(splots, print=TRUE, ncol=4, nrow=3)
```
